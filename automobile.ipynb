```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
data = pd.read_csv('Automobile+29.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.columns
```

```python
corr_matrix = data.corr(numeric_only = True)
```

```python
word_to_num = {
'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6,
    'eight': 8, 'twelve': 12, 'one': 1, 'zero': 0
}
```

```python
data['number_of_doors'] = data['number_of_doors'].map(word_to_num)
data['number_of_cylinders'] = data['number_of_cylinders'].map(word_to_num)
```

```python
data[['number_of_doors', 'number_of_cylinders']].dtypes, data[['number_of_doors', 'number_of_cylinders']].head()
```

```python
data.hist(figsize=(16, 14), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Distribution of Numeric Features", fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

```python
plt.figure(figsize = (10,6))
sbn.heatmap(corr_matrix, annot = True, cmap = "coolwarm", fmt = ".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sbn.histplot(data['make'], bins=30, )
plt.title("Distribution of cars by make")
plt.xlabel("Make")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sbn.histplot(data['price'], bins=30, kde=True)
plt.title("Distribution of Price")
plt.xlabel("price")
plt.ylabel("Count")
plt.show()
```

```python
sbn.histplot(data['engine_size'], bins=30, kde=True)
plt.title("Distribution of Engine size")
plt.xlabel("engine size")
plt.ylabel("Count")
plt.show()
```

```python
sbn.histplot(data['compression_ratio'], bins=30, kde=True)
plt.title("Distribution of Compression ratio")
plt.xlabel("compression ratio")
plt.ylabel("Count")
plt.show()
```

```python

categorical_columns = ['body_style', 'number_of_doors', 'drive_wheels', 'fuel_type', 'aspiration']

for col in categorical_columns:
    plt.figure(figsize=(10, 6))
    sbn.countplot(data=data, x=col, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
sbn.boxplot(x='fuel_type', y='price', data=data)
plt.title("Price distribution across fuel types")
plt.show()
```

```python
avgprice = data.groupby('make')['price'].mean()
avgprice.plot(kind='bar', figsize=(10,6), title='Average Car Price by the Make')
plt.ylabel('Average car Price')
plt.show()
```

```python
data['body_style'].value_counts().plot(kind='bar', title='Number of Cars per Body Style', color='green')
plt.ylabel('Count')
plt.xlabel('Body Style')
plt.show()
```

```python
avg_horsepower = data.groupby('number_of_cylinders')['horsepower'].mean()
avg_horsepower.plot(kind='bar', color='maroon', title='Average Horsepower by Number of Cylinders')
plt.ylabel('Horsepower')
plt.xlabel('Number of Cylinders')
plt.tight_layout()
plt.show()
```

```python
data.groupby('fuel_type')['price'].mean().plot(kind='bar', color='red', title='Average Price by Fuel Type')
plt.ylabel('Average Price')
plt.xlabel('Fuel Type')
plt.tight_layout()
plt.show()
```

```python
data.groupby('drive_wheels')['price'].mean().plot(kind='bar', color='orange', title='Average Price by Drive Wheels')
plt.ylabel('Average Price')
plt.xlabel('Drive Wheels')
plt.show()
```

```python
data.groupby('aspiration')['horsepower'].mean().plot(kind='bar', color='magenta', title='Average Horsepower by Aspiration Type')
plt.ylabel('Average Horsepower')
plt.xlabel('Aspiration')
plt.show()
```

```python
data.groupby('engine_location')['price'].mean().plot(kind='bar', title='Average Price by Engine Location', color='cyan')
plt.ylabel('Average Price')
plt.xlabel('Engine Location')
plt.show()
```

```python
data['fuel_system'].value_counts().plot(kind='bar', title='Count of Cars by Fuel System', color='beige')
plt.ylabel('Count')
plt.xlabel('Fuel System')
plt.show()
```

```python
sbn.lmplot(x='engine_size', y='curb_weight', data=data)
plt.title("Curb Weight against Engine Size")
plt.show()
```

```python
sbn.lmplot(x="engine_size", y="price", data=data)
plt.title("Price against Engine Size")
plt.show()
```

```python
sbn.lmplot( x='horsepower', y='price', data=data)
plt.title('Price and horsepower')
```

```python
sbn.lmplot(x='compression_ratio', y= 'engine_size', data=data)
plt.title('Engine Size and Copmression Ratio')
```
